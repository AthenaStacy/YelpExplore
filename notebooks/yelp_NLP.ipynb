```python
###NOW FOR A LITTLE NLP

stops_extra = ['http', 'www', 'yelp', 'com', 'biz']

def review_to_words( raw_review ):
	# Cleaning technique learned from Kaggle Bag of Words competition:
    # Function to convert a raw review to a string of words
    #
    #Remove HTML
    review_text = BeautifulSoup(raw_review).get_text() 
    
    #Remove non-letters        
    letters_only = re.sub("[^a-zA-Z]", " ", review_text)
    
    #Convert to lower case, split into individual words
    words = letters_only.lower().split()                             
 
    stops = set(stopwords.words("english"))                  

    #Remove stop words
    meaningful_words = [w for w in words if not w in stops and not w in stops_extra]   

    return( " ".join( meaningful_words ))   
```

```python
#Clean the reviews
reviews_clean=[]
nrev = reviews['text'].size
for i in range( 0, nrev ):
	reviews_clean.append(review_to_words(reviews['text'][i]))
	if(i%1000 == 0):
		print("Cleaned %d of %d so far\n" %(i, nrev))
```
